```python
import pandas as pd
import numpy as np
%matplotlib inline
pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 100)
```

```python
dados = pd.read_csv("../data/gas-prices-in-brazil/2004-2019.tsv", sep = '\t', usecols=['DATA FINAL', 'REGIÃO', 'ESTADO', 'PRODUTO', 'PREÇO MÉDIO REVENDA'])
```

```python
dados.head()
```

```python
from unidecode import unidecode

dados.columns = [unidecode(x.lower().replace(" ", "_")) for x in dados.columns]
```

```python
dados.head()
```

```python
dados.produto.value_counts()
```

```python
diesel = dados.query("produto == 'ÓLEO DIESEL'")
```

```python
diesel.head(100)
```

```python
diesel.query("estado == 'CEARA'").head(100)
```

```python
diesel = diesel.assign(data_final = pd.to_datetime(diesel.data_final))
```

```python
diesel.dtypes
```

# Validação

```python
train = diesel.query("data_final < '2011-01-01'")
test = diesel.query("data_final >= '2011-01-01'")
train.shape, test.shape
```

# Baseline

```python
train.assign(baseline = train.groupby(['estado'])['preco_medio_revenda'].shift(1)) \
        .query("estado == 'CEARA'") \
            .head(100)
```

```python
train = train.assign(baseline = train.groupby(['estado'])['preco_medio_revenda'].shift(1))
test = test.assign(baseline = test.groupby(['estado'])['preco_medio_revenda'].shift(1))
```

```python
display(train.head())
display(test.head())
```

```python
from sklearn.metrics import mean_squared_log_error

print(np.sqrt(mean_squared_log_error(train.preco_medio_revenda[train.baseline.notnull()], 
                                     train.baseline[train.baseline.notnull()])))

print(np.sqrt(mean_squared_log_error(test.preco_medio_revenda[test.baseline.notnull()], 
                                     test.baseline[test.baseline.notnull()])))
```

# Machine Learning

```python
train.plot(x='data_final', y='preco_medio_revenda');
```

```python
# Estacionarizando a variável alvo
train.assign(diff_preco_medio_revenda = train.groupby(['estado'])['preco_medio_revenda'].apply(lambda x: x.diff().shift(-1))) \
        .query("estado == 'CEARA'")
```

```python
train = train.assign(diff_preco_medio_revenda = train.groupby(['estado'])['preco_medio_revenda'].apply(lambda x: x.diff().shift(-1)))
test = test.assign(diff_preco_medio_revenda = test.groupby(['estado'])['preco_medio_revenda'].apply(lambda x: x.diff().shift(-1)))
```

```python
train.head()
```

```python
train.plot(x='data_final', y='diff_preco_medio_revenda');
```

```python
test.plot(x='data_final', y='diff_preco_medio_revenda');
```

```python
# Variáveis sazonais
train = train.assign(mes = train.data_final.dt.month, 
                     dia = train.data_final.dt.day, 
                     dia_da_semana = train.data_final.dt.weekday, 
                     dia_do_ano = train.data_final.dt.dayofyear,
                     semana_do_ano = train.data_final.dt.weekofyear)

test = test.assign(mes = test.data_final.dt.month, 
                   dia = test.data_final.dt.day, 
                   dia_da_semana = test.data_final.dt.weekday, 
                   dia_do_ano = test.data_final.dt.dayofyear,
                   semana_do_ano = test.data_final.dt.weekofyear)
```

```python
# Lag
train = train.assign(lag = train.groupby(['estado'])['preco_medio_revenda'].shift()).query("estado == 'DISTRITO FEDERAL'")
test = test.assign(lag = test.groupby(['estado'])['preco_medio_revenda'].shift()).query("estado == 'DISTRITO FEDERAL'")
```

```python
# média móvel
train = train.assign(media_movel_4sem = train.groupby(['estado'])['preco_medio_revenda'].rolling(4).mean().reset_index(level=0, drop=True))
test = test.assign(media_movel_4sem = test.groupby(['estado'])['preco_medio_revenda'].rolling(4).mean().reset_index(level=0, drop=True))
```
