```python
diesel_treino.head()
```

### Feature Engineering

```python
df_X_treino = pd.DataFrame(index=diesel_treino.index)
df_X_valid = pd.DataFrame(index=diesel_valid.index)
```

```python
# A variável alvo não é uma variável estácionária: média e variância mudam ao longo do tempo

diesel_treino.plot(x='DATA FINAL', y='PREÇO MÉDIO REVENDA');
```

```python
# Estacionarizando a variável alvo
# O novo valor alvo é o valor atual menos o valor anterior

def target_diff(x):
    return x.diff().shift(-1)

diesel_treino['DIFF PREÇO MÉDIO REVENDA'] = diesel_treino.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].apply(target_diff)

diesel_treino.plot(x='DATA FINAL', y='DIFF PREÇO MÉDIO REVENDA');
```

```python
df_X_treino['DIFF_PRECO_MEDIO_REVENDA_PROXIMA_SEMANA'] = diesel_treino.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].apply(target_diff)
df_X_valid['DIFF_PRECO_MEDIO_REVENDA_PROXIMA_SEMANA'] = diesel_valid.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].apply(target_diff)
```

```python
display(df_X_treino.head())
display(df_X_valid.head())
```

```python
# Variáveis sazonais

df_X_treino['MES'] = diesel_treino['DATA FINAL'].dt.month
df_X_treino['DIA'] = diesel_treino['DATA FINAL'].dt.day
df_X_treino['DIA_DA_SEMANA'] = diesel_treino['DATA FINAL'].dt.weekday
df_X_treino['DIA_DO_ANO'] = diesel_treino['DATA FINAL'].dt.dayofyear
df_X_treino['SEMANA_DO_ANO'] = diesel_treino['DATA FINAL'].dt.weekofyear

df_X_valid['MES'] = diesel_valid['DATA FINAL'].dt.month
df_X_valid['DIA'] = diesel_valid['DATA FINAL'].dt.day
df_X_valid['DIA_DA_SEMANA'] = diesel_valid['DATA FINAL'].dt.weekday
df_X_valid['DIA_DO_ANO'] = diesel_valid['DATA FINAL'].dt.dayofyear
df_X_valid['SEMANA_DO_ANO'] = diesel_valid['DATA FINAL'].dt.weekofyear
```

```python
display(df_X_treino.head())
display(df_X_valid.head())
```

```python
# Lag
df_X_treino['PRECO_MEDIO_REVENDA_ATUAL'] = diesel_treino['PREÇO MÉDIO REVENDA']
df_X_valid['PRECO_MEDIO_REVENDA_ATUAL'] = diesel_valid['PREÇO MÉDIO REVENDA']
```

```python
display(df_X_treino.head())
display(df_X_valid.head())
```

```python
# Diferença do Lag
df_X_treino['DIFF_PRECO_MEDIO_REVENDA_ATUAL'] = diesel_treino.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].diff()
df_X_valid['DIFF_PRECO_MEDIO_REVENDA_ATUAL'] = diesel_valid.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].diff()
```

# Sequential Forward Feature Selection

```python
def run_randomForest(Xtrain, ytrain, Xval, yval, cols = None):
    model = RandomForestRegressor(n_jobs=-1, random_state=0, n_estimators=500)
    model.fit(Xtrain[cols], ytrain)
    p = model.predict(Xval[cols])
    
    p_final = Xval['PRECO_MEDIO_REVENDA_ATUAL'] + p
    yval_final = Xval['PRECO_MEDIO_REVENDA_ATUAL'] + yval
    
    erro = np.sqrt(mean_squared_log_error(yval_final, p_final)) * 100
    
    return erro
```

```python
def sequential_forward_feature_selection(Xtrain, ytrain, Xval, yval, cols_selected=[], run_model=run_randomForest):
    var_menor_erro = None
    valor_menor_erro = 1000.0
    
    for var in [x for x in Xtrain.columns if x not in cols_selected]:
        erro = run_model(Xtrain, ytrain, Xval, yval, cols=cols_selected+[var])
        
        print(f"Variáveis: {cols_selected+[var]} - Erro: {erro:.4f}\n")
        
        if erro < valor_menor_erro:
            var_menor_erro = var
            valor_menor_erro = erro
        
    print(f"Melhor Variável: {cols_selected+[var_menor_erro]} - Erro: {valor_menor_erro:.4f}\n")
    return cols_selected + [var_menor_erro]
```

```python
selected_feat1 = sequential_forward_feature_selection(Xtrain, ytrain, Xval, yval)
```

```python
selected_feat2 = sequential_forward_feature_selection(Xtrain, ytrain, Xval, yval, cols_selected=selected_feat1)
```

```python
selected_feat3 = sequential_forward_feature_selection(Xtrain, ytrain, Xval, yval, cols_selected=selected_feat2)
```

Como ao incrementar em cima da primeira variável selecionada não temos ganhos de performance, o nosso esquema de feature selection seleciona apenas a uma variável: `MES`
